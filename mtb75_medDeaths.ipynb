## Dataset Overview - Median Age of Death

I do not know about you all, but personally I want to life a long happy life, especially with a family. The longer family can spent time with each other the better. Therefore, I am analyzing the median age of death in each of the neighborhoods. In order for a neighborhood to be deemed the best, it must have the largest median age of death.

In this notebook, I will going over three different sets of data:
1. Median age of black deaths in the different neighborhoods
2. Median age of white deaths in the different neighborhoods
3. Total median age of deaths in the different neighborhoods

# Lets Begin:

```python
#Calling in the data set and dropping missing values.
deaths = pd.read_csv("deaths.csv")
deaths = deaths.dropna()
deaths = deaths.drop(deaths.index[0])
```

# Median Age of Black Deaths

To begin, I made a density plot to visualize the approximately normal distribution of deaths.

The distribution of deaths appears to be uniform, with the median age of death to be 68.45.

```python
deaths['Black MD AGE AT DEATH'].plot.density()
print("The black median age of death in Pittsburgh is", deaths['Black MD AGE AT DEATH'].median())
```

# Bringing the Data to the Neighborhoods

All the information is great, but the goal is to find the best neighborhood to raise a family which, of course, living awhile would help with that. So, with a bar chart, we can find the neighborhood with the longest median lifespan of a black person.

With this we get our top 5:

1. Squirrel Hill South
2. Shadyside
3. Glen Hazel
4. Upper Hill
5. Larimer

With this, we can conclude that Squirrel Hill South has the longest median lifespan for a black person at 83.2 years of age.

```python
b_med_deaths = deaths.groupby("NEIGHBORHOOD").sum()['Black MD AGE AT DEATH']
b_med_deaths = b_med_deaths.sort_values(ascending = False).head(5)
b_med_deaths.plot(kind='bar', x='NEIGHBORHOOD', y='Black MD AGE AT DEATH', figsize=(30,20))
print("The largest value is", deaths['Black MD AGE AT DEATH'].max())
```
